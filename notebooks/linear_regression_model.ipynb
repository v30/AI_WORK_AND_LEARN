```python
# Import libs
import numpy as np
import matplotlib.pyplot as plt

# Create original data
day = np.array([1, 2, 3, 4, 5, 6, 7])
temp = np.array([30, 32, 31, 29, 28, 27, 26])

# fit the regression line, polyfit returns the slope (m) and intercept (b)
m, b = np.polyfit(day, temp, 1)

# predict the new values
new_days = np.array([8, 9])
predicted_temps = m * new_days + b

# plot original data and regression line
plt.scatter(day, temp, color='blue', label='Original Data')
plt.plot(day, m * day + b, color='red', label='Regression Line')

# plot new data points
plt.scatter(new_days, predicted_temps, color='green', label='Predicted Data')

# optional: Annotate the predicted points
for price, rooms in zip(new_days, predicted_temps):
    plt.text(price, rooms + 0.1, f'{rooms:.2f} rooms', ha='center', color='green')

# add labels and title
# sets the label for the x
plt.xlabel('Day')
# sets the label for the y
plt.ylabel('Temperature (°C)')
# sets the title for the plot
plt.title('Temperature Prediction using Linear Regression')
# add legend and grid
plt.legend()
plt.grid(True)
# adjust layout to prevent overlap
plt.tight_layout()

# show the plot
plt.show()
```
