```python
import numpy as np
import matplotlib.pyplot as plt

# Get the data
data = np.loadtxt('../data/auto-mpg.csv', delimiter=',', usecols=(0, 3), skiprows=1)
mpg = np.array(data[:, 0])
hp = np.array(data[:, 1])

# fit a linear regression model
m, b = np.polyfit(hp, mpg, 1)

# make predictions
pred_hp = np.array([103, 142, 164, 197])
predicted_mpg = m * pred_hp + b

print(f'the slope is {m:.2f} and the intercept is {b:.2f}')
print(f'predicted mpg for hp {pred_hp} is {predicted_mpg}')

# plot original data
plt.scatter(hp, mpg, label='Data', color='cyan')
plt.plot(hp, m * hp + b, color='blue', label='Fitted Line')

# plot new data points
plt.scatter(pred_hp, predicted_mpg, color='red', label='Predictions')

# optional annotations
for hp, mpgs in zip(pred_hp, predicted_mpg):
    plt.text(hp, mpgs, f'({mpgs:.2f}) BHP', fontsize=12, ha='center', color='red')

plt.title('MPG vs Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Miles Per Gallon (MPG)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
