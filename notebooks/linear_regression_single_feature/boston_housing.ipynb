```python
# imports
import numpy as np
import matplotlib.pyplot as plt

# prep data
data = np.loadtxt('../../data/Boston.csv', delimiter=',', usecols=(1, 7), skiprows=1)

crim = np.array(data[:, 0])
age = np.array(data[:, 1])

# fit to linear regression line
m, b = np.polyfit(age, crim, 1)

# make predictions
new_ages = np.array([21, 45, 65, 89]);
predictions = m * new_ages + b;

# plot original data
plt.scatter(age, crim, label = 'Data Points', color='blue')
plt.plot(age, m * age + b, color='red', label='Regression line')

# plot new data points
plt.scatter(new_ages, predictions, color='green', label='Predicted Crimes')

# annotations for the new data points
for crim, ages in zip(new_ages, predictions):
    plt.text(crim, ages + 0.1, f'{ages:.2f}', ha='center', fontsize= 12, color='green')

plt.title('Age vs Crim')
plt.xlabel('Age')
plt.ylabel('Crim')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()



```
