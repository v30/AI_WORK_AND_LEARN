```python
# imports
import numpy as np
import matplotlib.pyplot as plt

# get and prep data
data = np.loadtxt('../../data/advertising.csv', delimiter=',', usecols=(0, 3), skiprows=1)

tv = np.array(data[:, 0])
sales = np.array(data[:, 1])

# fit a polynomial linear regression model
m, b = np.polyfit(tv, sales, 1)
print(f'slope is {m} and intercept is {b}')

# make predictions
new_tv_adds = np.array([0, 100, 132, 175, 183, 200])
predictions = m * new_tv_adds + b

# plot original data
plt.scatter(tv, sales, color='blue', label='Original Data')
# plot regression line
plt.plot(tv, m * tv + b, color='red', label='Fitted Line')

# plot predictions
plt.scatter(new_tv_adds, predictions, color='green', label='Predictions')

# add annotations
for tv, ads in zip(new_tv_adds, predictions):
    plt.text(tv, ads, f'({ads:.2f})', fontsize=12, ha='center', color='green')

# legend and labels
plt.legend()
plt.xlabel('TV Advertising Budget (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.title('TV Advertising vs Sales with Predictions')
plt.grid(True)
plt.tight_layout# show plot
plt.show()

# get and prep data
data = np.loadtxt('../../data/advertising.csv', delimiter=',', usecols=(2, 3), skiprows=1)

news_pap = np.array(data[:, 0])
sales = np.array(data[:, 1])

# fit a polynomial linear regression model
m, b = np.polyfit(news_pap, sales, 1)
print(f'slope is {m} and intercept is {b}')

# make predictions
new_news_pap = np.array([27, 52, 90])
predictions = m * new_news_pap + b

# plot original data
plt.scatter(news_pap, sales, color='blue', label='Original Data')
# plot regression line
plt.plot(news_pap, m * news_pap + b, color='red', label='Fitted Line')

# plot predictions
plt.scatter(new_news_pap, predictions, color='green', label='Predictions')

# add annotations
for news_pap, ads in zip(new_news_pap, predictions):
    plt.text(news_pap, ads, f'({ads:.2f})', fontsize=12, ha='center', color='green')

# legend and labels
plt.legend()
plt.xlabel('news_pap Advertising Budget (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.title('news_pap Advertising vs Sales with Predictions')
plt.grid(True)
plt.tight_layout# show plot
plt.show()

# get and prep data
data = np.loadtxt('../../data/advertising.csv', delimiter=',', usecols=(1, 3), skiprows=1)

radio = np.array(data[:, 0])
sales = np.array(data[:, 1])

# fit a polynomial linear regression model
m, b = np.polyfit(radio, sales, 1)
print(f'slope is {m} and intercept is {b}')

# make predictions
new_radio_adds = np.array([0, 15, 17, 20, 26])
predictions = m * new_radio_adds + b

# plot original data
plt.scatter(radio, sales, color='blue', label='Original Data')
# plot regression line
plt.plot(radio, m * radio + b, color='red', label='Fitted Line')

# plot predictions
plt.scatter(new_radio_adds, predictions, color='green', label='Predictions')

# add annotations
for radio, ads in zip(new_radio_adds, predictions):
    plt.text(radio, ads, f'({ads:.2f})', fontsize=12, ha='center', color='green')

# legend and labels
plt.legend()
plt.xlabel('radio Advertising Budget (in thousands)')
plt.ylabel('Sales (in thousands)')
plt.title('radio Advertising vs Sales with Predictions')
plt.grid(True)
plt.tight_layout# show plot
plt.show()
```
