```python
# import libs
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get some data, in this case will use one of the existing sets
# the np.loadtxt is expecting all values to be numeric
# data = np.loadtxt("../../data/Salary_Data.csv", delimiter=',', skiprows=1)

# This is better to load data that is not only numeric
data = np.genfromtxt('../../data/Salary_Data.csv', delimiter=',', dtype=None, encoding='utf-8', names=True)

# When you use np.genfromtxt() with names=True, it returns a structured array. That means instead of accessing data with numerical indexing like data [:, 0], you access columns by name
# ages = data[:, 0]
# salaries = data[:, 5]
# get the name of the columns like this:
# print(data.dtype.names)
ages = np.array(data['\ufeffAge'])
salaries = np.array(data['Salary'])
experience = np.array(data['Years_of_Experience'])
```

```python
# Match to a plot line
m, b = np.polyfit(ages, salaries, 1)
```

```python
# make predictions
new_ages = np.array([25, 45, 55])
predicted = m * new_ages + b
```

```python
# plot the original data and linear regression plot line
plt.scatter(ages, salaries, label='Data Points', color='blue')
plt.plot(ages, m * ages + b, color='red', label='Regression Line')

# plot prediction
plt.scatter(new_ages, predicted, color='green', label='Predicted Salaries')

# annotate the plot point
for ages, salary in zip(new_ages, predicted):
    plt.text(ages, salary + 0.1, f'{salary:.2f} points', ha='center', color='green')

# add labels and title
plt.xlabel('Age in years')
plt.ylabel('Salary in $')
plt.title('Predicting Salary Based on Age')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print(new_ages)
for i in range(len(new_ages)):
    print(f'If you are {new_ages[i]} years old your salary on average will be ${predicted[i]:.2f}')
```

```python
# match a new plot line to the salary vs experience column
m, b = np.polyfit(experience, salaries, 1)
```

```python
# make predictions
exp_years = np.array([2, 5, 10])
predicted = m * exp_years + b
print(predicted)
```

```python
# plot the original data and linear regression plot line
plt.scatter(experience, salaries, label='Data Points', color='blue')
plt.plot(experience, m * experience + b, color='red', label='Regression Line')

# plot prediction
plt.scatter(exp_years, predicted, color='green', label='Predicted Salaries')

# annotate the plot point
for experience, salary in zip(exp_years, predicted):
    plt.text(experience, salary + 0.1, f'{salary:.2f} points', ha='center', color='green')

# add labels and title
plt.xlabel('Experience in years')
plt.ylabel('Salary in $')
plt.title('Predicting Salary Based on Experience')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
