```python
# imports
import numpy as np
import matplotlib.pyplot as plt

# dummy data
hours = np.array([1, 2, 3, 4, 5, 10, 11, 12, 13, 14, 15])
score = np.array([8, 9, 12, 14, 17, 66, 78, 86, 92, 112, 123])

# fit the linear regression line
m, b = np.polyfit(hours, score, 1)

# predict scores using the linear regression line
new_hours = np.array([0, 7, 9, 18])
predicted_scores = m * new_hours + b

# plot original data and regression line
plt.scatter(hours, score, color='blue', label='Original Data')
plt.plot(new_hours, predicted_scores, color='red', label='Regression Line')

# plot new data points
plt.scatter(new_hours, predicted_scores, color='green', label='Predicted Scores')

# annotate the plot
for hours, scores in zip(new_hours, predicted_scores):
    plt.text(hours, scores + 0.1, f'{scores:.2f} points', ha='center', color='green')

# add labels and title
plt.xlabel('Practice Hours')
plt.ylabel('Score in Points (out of 150)')
plt.title('Predicting Scores Based on Practice Hours')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
