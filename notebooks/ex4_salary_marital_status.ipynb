```python
import sys
import os

sys.path.append(os.path.abspath('../utils'))
from create_data import create_dummy_study_data
import numpy as np
import matplotlib.pyplot as plt

# Create dummy data
data = create_dummy_study_data("../data/salary_marital.csv", 42, 250, [0, 1], [15, 100])

# Preprocess data
data = np.loadtxt('../data/salary_marital.csv', delimiter=',', skiprows=1)
salary = np.array(data[:, 0])
marital_status = np.array(data[:, 1])

# Fit to linear regression model
m, b = np.polyfit(salary, marital_status, 1)

# Make predictions
prediction_salaries = np.array([13])
predictions = m * prediction_salaries + b

# plot original data
plt.scatter(salary, marital_status, color='blue', label='Data Points')
plt.plot(salary, m * salary + b, color='red', label='Fitted Line')

plt.scatter(prediction_salaries, predictions, color='green', label='Predictions', marker='x')

# optional annotations
for salary, statuses in zip(prediction_salaries, predictions):
    plt.text(salary, statuses, f'({statuses:.2f}) POINTS', fontsize=12, ha='center', color='green')

plt.title('Salary vs Marital Status')
plt.xlabel('Salary')
plt.ylabel('Marital Status')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
